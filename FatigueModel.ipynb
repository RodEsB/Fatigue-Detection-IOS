##Data pipeline preprocessing functions
In this section, the data generator (data_generator) was built. This function serves as the "engine" of the pipeline, responsible for reading images from the disk, processing them into batches, and efficiently serving them to the model. This allows the model to be trained on thousands of images without saturating RAM.

```python
import os
import shutil
import numpy as np

# Path to the source DDD image dataset
#ddd_source_path = '/root/.cache/kagglehub/datasets/ismailnasri20/driver-drowsiness-dataset-ddd/versions/1/Driver Drowsiness Dataset (DDD)'
ddd_source_path = '/kaggle/input/driver-drowsiness-dataset-ddd/Driver Drowsiness Dataset (DDD)'
# Main project directory we will create
base_dir = 'drowsiness_project_opencv'

# Directories for our unified and split data
unified_dir = os.path.join(base_dir, 'unified_dataset')
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

# Ratios for splitting (70% train, 15% validation, 15% test)
train_ratio = 0.70
validation_ratio = 0.15


print("--- Step 1: Preparing and Splitting Data for Drowsy/Non-Drowsy ---")

# Clean up previous runs by deleting the main project directory if it exists
if os.path.exists(base_dir):
    shutil.rmtree(base_dir)

# --- Add verification step for the ddd_source_path ---
print(f"\nVerifying the source dataset path: {ddd_source_path}")
if os.path.exists(ddd_source_path):
    print("Source path exists. Listing contents:")
    try:
        for item in os.listdir(ddd_source_path):
            print(f"- {item}")
    except Exception as e:
        print(f"Error listing contents: {e}")
else:
    print("Source path does NOT exist. Please check the download path.")
    pass

print("\n--> Unifying source dataset into a binary (drowsy/non_drowsy) structure...")
# Mapping directly from main categories to unified classes
ddd_map = {
    'Drowsy': 'drowsy',
    'Non Drowsy': 'non_drowsy'
}
final_classes = ['drowsy', 'non_drowsy']


os.makedirs(unified_dir, exist_ok=True)
for cls in final_classes:
    os.makedirs(os.path.join(unified_dir, cls), exist_ok=True)

def copy_images(src_dir, dest_dir):
    if os.path.exists(src_dir): # Add check if source directory exists
        for img_file in os.listdir(src_dir):
            if os.path.isfile(os.path.join(src_dir, img_file)):
                shutil.copy(os.path.join(src_dir, img_file), dest_dir)
    else:
        print(f"Source directory does not exist: {src_dir}. Skipping copy.") # Add message if source directory doesn't exist

# Iterate through the main categories (Drowsy, Non Drowsy)
for main_category, unified_class in ddd_map.items():
    src_folder = os.path.join(ddd_source_path, main_category)
    dest_folder = os.path.join(unified_dir, unified_class)
    copy_images(src_folder, dest_folder)


print("Dataset unification process attempted. Check logs for missing source directories.")

print("\n--> Splitting unified data into train, validation, and test sets...")
for cls in final_classes:
    # Create the destination subdirectories
    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(validation_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

    # Get a shuffled list of all files for the current class
    src_folder = os.path.join(unified_dir, cls)
    all_files = os.listdir(src_folder)
    np.random.shuffle(all_files)

    # Calculate split indices
    train_split_index = int(len(all_files) * train_ratio)
    validation_split_index = int(len(all_files) * (train_ratio + validation_ratio))

    # Slice the list of files
    train_files = all_files[:train_split_index]
    validation_files = all_files[train_split_index:validation_split_index]
    test_files = all_files[validation_split_index:]

    # Copy the files to their final destinations
    for f in train_files: shutil.copy(os.path.join(src_folder, f), os.path.join(train_dir, cls, f))
    for f in validation_files: shutil.copy(os.path.join(src_folder, f), os.path.join(validation_dir, cls, f))
    for f in test_files: shutil.copy(os.path.join(src_folder, f), os.path.join(test_dir, cls, f))

print("Dataset splitting process attempted. Check verification counts.")

# Verification
print("\n--- Verification: Final file counts ---")
for cls in final_classes:
    train_count = len(os.listdir(os.path.join(train_dir, cls)))
    val_count = len(os.listdir(os.path.join(validation_dir, cls)))
    test_count = len(os.listdir(os.path.join(test_dir, cls)))
    print(f" -> Class '{cls}': {train_count} train, {val_count} val, {test_count} test.")
```

###Filters for image processing
Here, the preprocess_image function was defined, which uses the OpenCV library to apply a series of transformations to each image. Optional filters were implemented for noise reduction (Median Blur), contrast enhancement (CLAHE), and edge detection (Canny), in addition to data augmentation techniques to make the model more robust.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

IMG_WIDTH, IMG_HEIGHT = 100, 100

def preprocess_image(image_path, augment=False, use_canny=False):
    """
    Loads an image and applies advanced preprocessing, augmentation, and edge detection.
    """
    # Load image in grayscale using OpenCV
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Warning: Could not read image at {image_path}")
        return None

    # Resize image to the target dimensions
    image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT))

    # Noise Reduction: Median Blur preserves edges well.
    image = cv2.medianBlur(image, 3)

    # Contrast Enhancement: CLAHE is excellent for varying lighting.
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    image = clahe.apply(image)

    # Data Augmentation block (applied before Canny)
    if augment:
        if np.random.rand() > 0.5: image = cv2.flip(image, 1)
        if np.random.rand() > 0.75:
            angle = np.random.uniform(-15, 15)
            M = cv2.getRotationMatrix2D((IMG_WIDTH / 2, IMG_HEIGHT / 2), angle, 1)
            image = cv2.warpAffine(image, M, (IMG_WIDTH, IMG_HEIGHT))
        if np.random.rand() > 0.75:
            value = np.random.uniform(-25, 25)
            image = np.clip(image.astype(np.int16) + value, 0, 255).astype(np.uint8)

    # Edge Detection: Canny
    if use_canny:
        image = cv2.Canny(image, 50, 150)

    # Rescale pixels from [0, 255] to [0.0, 1.0]
    image = image.astype('float32') / 255.0

    # Expand dimensions to (width, height, 1)
    return np.expand_dims(image, axis=-1)

print("OpenCV preprocessing function 'preprocess_image' (V2) is defined with new filters.")
```

###Personalized data
This section describes the initial process of data preparation and splitting. Images from the original dataset were taken, the classes were unified and standardized into drowsy and non_drowsy, and finally, the dataset was split into three parts: training, validation, and test. This ensures a rigorous evaluation methodology and reliable results.

```python
import os
import numpy as np
import tensorflow as tf

BATCH_SIZE = 32


def data_generator(directory, batch_size, class_indices, augment=False, use_blur=False, use_clahe=False, use_canny=False):
    """
    Corrected generator that yields integer labels for binary classification.
    """
    filepaths = []
    labels = []
    class_names = sorted(class_indices.keys())

    for class_name in class_names:
        class_idx = class_indices[class_name]
        class_path = os.path.join(directory, class_name)
        if os.path.isdir(class_path):
            for f in os.listdir(class_path):
                filepaths.append(os.path.join(class_path, f))
                labels.append(class_idx)

    filepaths = np.array(filepaths)
    labels = np.array(labels)

    while True:
        indices = np.arange(len(filepaths))
        np.random.shuffle(indices)

        for i in range(0, len(indices), batch_size):
            batch_indices = indices[i : i + batch_size]

            if len(batch_indices) < batch_size and i > 0:
                continue

            batch_paths = filepaths[batch_indices]
            batch_labels_int = labels[batch_indices]
            #, use_blur=use_blur, use_clahe=use_clahe, use_canny=use_canny
            batch_images = []
            for path in batch_paths:
                processed_img = preprocess_image(path, augment=augment)
                if processed_img is not None:
                    batch_images.append(processed_img)

            if len(batch_images) == 0:
                continue

            yield np.array(batch_images), batch_labels_int

print("Corrected data generator function 'data_generator' is defined.")
```

##Building the model architecture
Here, the architecture of the Convolutional Neural Network (CNN) is defined. The structure of the layers is detailed, including convolutional layers (Conv2D) for visual feature extraction, MaxPooling2D layers to reduce dimensionality, and dense layers (Dense) that perform the final classification between drowsy and non_drowsy.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

IMG_WIDTH, IMG_HEIGHT = 300, 300

NUM_CLASSES = 2

print("--- Step 4: Building the CNN Model ---")

# The input shape for the first layer
input_shape = (IMG_WIDTH, IMG_HEIGHT, 1) # 1 for grayscale

model = Sequential([
    # Block 1
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPooling2D((2, 2)),

    # Block 2
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Block 3
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Flatten the results to feed into a dense layer
    Flatten(),

    # A dense layer for classification
    Dense(128, activation='relu'),

    # Dropout layer to help prevent overfitting
    Dropout(0.5),

    # Final output layer for BINARY classification
    # It has 1 neuron and a sigmoid activation function
    Dense(1, activation='sigmoid')
])

# For binary classification, we use 'binary_crossentropy' as the loss function
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Print a summary of the model you just built
model.summary()
```

###Training and Model evaluation
This section covers the model training process (model.fit), where the neural network learns from the training data. Subsequently, a performance evaluation is carried out using the test set (data the model has never seen before) to calculate key metrics such as accuracy and loss.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

EPOCHS = 20
BATCH_SIZE = 32

# Directories and class info needed for the generators
base_dir = 'drowsiness_project_opencv'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

class_names = sorted(os.listdir(train_dir))
class_indices = {name: i for i, name in enumerate(class_names)}

print("--- Step 5a: Training the Model ---")

train_gen = data_generator(train_dir, BATCH_SIZE, class_indices, augment=True, use_canny=False)
validation_gen = data_generator(validation_dir, BATCH_SIZE, class_indices, augment=False, use_canny=False)

num_train_samples = sum([len(files) for r, d, files in os.walk(train_dir)])
num_val_samples = sum([len(files) for r, d, files in os.walk(validation_dir)])
steps_per_epoch = num_train_samples // BATCH_SIZE
validation_steps = num_val_samples // BATCH_SIZE

history = model.fit(
    train_gen,
    steps_per_epoch=steps_per_epoch,
    epochs=EPOCHS,
    validation_data=validation_gen,
    validation_steps=validation_steps
)
print("\nModel training complete.")

print("\n--- Step 5b: Evaluating Model Performance ---")

# Plotting code remains the same
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Evaluate the model on the unseen test set
print("\n--> Evaluating on the test set...")
test_gen = data_generator(test_dir, BATCH_SIZE, class_indices, augment=False, use_canny=False)
num_test_samples = sum([len(files) for r, d, files in os.walk(test_dir)])
test_steps = num_test_samples // BATCH_SIZE if num_test_samples >= BATCH_SIZE else 1

loss, accuracy = model.evaluate(test_gen, steps=test_steps)
print(f"\nTest Accuracy: {accuracy*100:.2f}%")
print(f"Test Loss: {loss:.4f}")

print("\n--> Generating Classification Report and Confusion Matrix...")
test_gen_for_report = data_generator(test_dir, BATCH_SIZE, class_indices, augment=False, use_canny=False)

# Get true labels (they are now integers, not one-hot)
y_true = []
num_batches = 0
for _, labels_batch in test_gen_for_report:
    y_true.extend(labels_batch) # Simply extend the list with the integer labels
    num_batches += 1
    if num_batches >= test_steps:
        break

# Get predictions from the model
predictions = model.predict(test_gen_for_report, steps=test_steps)
# For a sigmoid output, we round the probability to get the final class (0 or 1)
y_pred = (predictions > 0.5).astype("int32").flatten()

# Ensure lengths of true labels and predictions match
y_true = np.array(y_true[:len(y_pred)])

# Classification Report
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# Confusion Matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

###Experimentation and comparision of results
Finally, an ablation study was implemented to measure the impact of the different OpenCV filters. A function was created to run multiple experiments automatically, training a model for each filter configuration. The accuracy results from each experiment were collected and presented in a comparative bar chart to objectively determine which preprocessing technique was most effective.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

IMG_WIDTH, IMG_HEIGHT = 300, 300

def preprocess_image(image_path, augment=False, use_blur=False, use_clahe=False, use_canny=False):
    """
    Loads an image and applies a specific combination of preprocessing filters.
    """
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None: return None

    image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT))

    # These filters will only be applied if their corresponding flag is True
    if use_blur:
        image = cv2.medianBlur(image, 3)

    if use_clahe:
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        image = clahe.apply(image)

    # Augmentation is applied before Canny
    if augment:
        if np.random.rand() > 0.5: image = cv2.flip(image, 1)
        if np.random.rand() > 0.75:
            angle = np.random.uniform(-15, 15)
            M = cv2.getRotationMatrix2D((IMG_WIDTH / 2, IMG_HEIGHT / 2), angle, 1)
            image = cv2.warpAffine(image, M, (IMG_WIDTH, IMG_HEIGHT))

    if use_canny:
        # Canny works best on a clean image, so we apply it after blur/clahe
        image = cv2.Canny(image, 50, 150)

    # Rescale pixels and expand dimensions
    image = image.astype('float32') / 255.0
    return np.expand_dims(image, axis=-1)

print("Flexible preprocessing function 'preprocess_image' (V3) is defined.")
```

```python

from tensorflow.keras.models import clone_model
import time

def run_experiment(model_to_clone, experiment_name, use_blur=False, use_clahe=False, use_canny=False):
    """
    Builds, trains, and evaluates a model with a specific preprocessing configuration.
    """
    print(f"\n{'='*25}")
    print(f"RUNNING EXPERIMENT: {experiment_name}")
    print(f"Configuration: Blur={use_blur}, CLAHE={use_clahe}, Canny={use_canny}")
    print(f"{'='*25}\n")

    # Start with a fresh, untrained model for a fair comparison
    fresh_model = clone_model(model_to_clone)
    fresh_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Create data generators with the specified configuration
    train_gen = data_generator(train_dir, BATCH_SIZE, class_indices, augment=True,
                               use_blur=use_blur, use_clahe=use_clahe, use_canny=use_canny)
    validation_gen = data_generator(validation_dir, BATCH_SIZE, class_indices, augment=False,
                                    use_blur=use_blur, use_clahe=use_clahe, use_canny=use_canny)
    test_gen = data_generator(test_dir, BATCH_SIZE, class_indices, augment=False,
                              use_blur=use_blur, use_clahe=use_clahe, use_canny=use_canny)

    # Train the model
    start_time = time.time()
    fresh_model.fit(
        train_gen,
        steps_per_epoch=steps_per_epoch,
        epochs=EPOCHS,
        validation_data=validation_gen,
        validation_steps=validation_steps,
        verbose=0 # Keep the output clean
    )
    end_time = time.time()
    print(f"Training complete in {end_time - start_time:.2f} seconds.")

    # Evaluate and print results
    print("--> Evaluating on the test set...")
    loss, accuracy = fresh_model.evaluate(test_gen, steps=test_steps, verbose=0)

    print(f"\n--- RESULTS FOR '{experiment_name}' ---")
    print(f"Test Accuracy: {accuracy*100:.2f}%")
    print("-" * 30)

    return { "name": experiment_name, "accuracy": accuracy }

print("Experiment runner function 'run_experiment' is defined.")
```

```python
import os
import cv2
import numpy as np

# Define Paths
# Based on your previous output, this is the path
BASE_DATA_PATH = "/root/.cache/kagglehub/datasets/enider/yawdd-dataset/versions/4"

# Define the source directories
SOURCE_DIRS = [
    os.path.join(BASE_DATA_PATH, "Mirror/Mirror/Female_mirror"),
    os.path.join(BASE_DATA_PATH, "Mirror/Mirror/Male_mirror Avi Videos")
]

# !<--- USER: Define a path for the NEW, organized dataset of images
NEW_DATA_DIR = "/content/YAWDD_Organized_Frames"

# Define our final class names
CLASSES = ["no_yawn", "yawn"]

# Define the keywords we will look for in the filenames
# These are based on the 'Action' column from Table1.pdf [cite: 27]
LABEL_KEYWORDS = {
    "yawn": ["Yawn"],
    "no_yawn": ["Talk", "Normal"]
}

# Create the New Directory Structure
print(f"Creating new directory structure at: {NEW_DATA_DIR}")
for class_name in CLASSES:
    class_path = os.path.join(NEW_DATA_DIR, class_name)
    os.makedirs(class_path, exist_ok=True) # exist_ok=True prevents errors

# Frame Extraction and AUTOMATIC Labeling
print("Starting frame extraction from 'Mirror' dataset...")

IMG_HEIGHT = 64
IMG_WIDTH = 64
frame_count = 0
unknown_files = []

for source_dir in SOURCE_DIRS:
    if not os.path.isdir(source_dir):
        print(f"Warning: Directory not found, skipping: {source_dir}")
        continue

    print(f"Processing directory: {source_dir}")

    for filename in os.listdir(source_dir):
        # We only care about .avi files
        if not filename.endswith(".avi"):
            continue

        # Automatic Labeling Logic 
        label = None

        # Check for 'yawn' keywords
        for keyword in LABEL_KEYWORDS["yawn"]:
            if keyword in filename:
                label = "yawn"
                break

        # If not a yawn, check for 'no_yawn' keywords
        if label is None:
            for keyword in LABEL_KEYWORDS["no_yawn"]:
                if keyword in filename:
                    label = "no_yawn"
                    break

        # End of Labeling Logic

        # If we found a label for this file, process it
        if label:
            # Define the full path to the video
            video_path = os.path.join(source_dir, filename)

            # Define the destination folder for the frames
            dest_folder = os.path.join(NEW_DATA_DIR, label)

            # Start Video Processing (OpenCV)
            try:
                cap = cv2.VideoCapture(video_path)
                if not cap.isOpened():
                    print(f"Error: Could not open video file: {video_path}")
                    continue

                frame_idx = 0
                while True:
                    ret, frame = cap.read()
                    # If 'ret' is False, we reached the end of the video
                    if not ret:
                        break

                    # Pre-process the frame
                    img_resized = cv2.resize(frame, (IMG_WIDTH, IMG_HEIGHT))

                    # Create a unique filename for the frame
                    frame_filename = f"{os.path.splitext(filename)[0]}_frame_{frame_idx:04d}.jpg"
                    save_path = os.path.join(dest_folder, frame_filename)

                    # Save the resized frame as a JPG
                    cv2.imwrite(save_path, img_resized)

                    frame_idx += 1
                    frame_count += 1

                cap.release()
                if frame_idx > 0:
                    print(f"  > Extracted {frame_idx} frames from {filename} -> {label}")

            except Exception as e:
                print(f"Error processing video {video_path}: {e}")

        # If we didn't find a label, log it as unknown
        else:
             if filename not in unknown_files:
                unknown_files.append(filename)

print("\n--- Frame Extraction Complete! ---")
print(f"Total frames extracted and saved: {frame_count}")
if unknown_files:
    print(f"\nWarning: {len(unknown_files)} files were found but could not be auto-labeled.")
    print("This is normal for files that are not 'Yawn', 'Talk', or 'Normal'.")
    print("Example unknown files:", unknown_files[:10]) # Show first 10

print(f"\nYour dataset of frames is ready at: {NEW_DATA_DIR}")
```

##CNN and LSTM Model Training

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, LSTM, TimeDistributed
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.model_selection import train_test_split

# Global Parameters
IMG_HEIGHT = 64
IMG_WIDTH = 64
CHANNELS = 3
SEQ_LENGTH = 20  # Number of frames in one sequence
BATCH_SIZE = 32
EPOCHS = 20      # Reduced for a faster test run

DATA_DIR = "/content/YAWDD_Organized_Frames"
CLASSES = ["no_yawn", "yawn"] # Class 0, Class 1

# Reduced to 12.5k per class (25k total frames)
MAX_FRAMES_PER_CLASS = 12500

# --- 2. Data Loading Function ---
def load_frames_from_folders(data_dir, classes, img_height, img_width, max_frames_per_class):
    """
    Loads a *subset* of frames from subfolders into two numpy arrays.
    """
    print(f"Loading data from {data_dir}...")
    X = []  # To store frames
    y = []  # To store labels

    for label_index, class_name in enumerate(classes):
        class_path = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_path):
            print(f"Error: Directory not found {class_path}")
            continue

        print(f"Loading frames from: {class_path} (Label: {label_index})")

        image_names = os.listdir(class_path)
        frame_counter = 0

        for img_name in image_names:
            if frame_counter >= max_frames_per_class:
                print(f"  > Reached limit of {max_frames_per_class} frames for class '{class_name}'.")
                break

            img_path = os.path.join(class_path, img_name)
            try:
                img = cv2.imread(img_path)
                if img is None: continue

                img_resized = cv2.resize(img, (img_width, img_height))
                img_normalized = img_resized / 255.0

                X.append(img_normalized)
                y.append(label_index)
                frame_counter += 1

            except Exception as e:
                print(f"Error processing image {img_path}: {e}")

    print(f"Finished loading data. Total frames: {len(X)}")

    X = np.array(X)
    y = np.array(y)

    return X, y


# Custom Data Generator
class SequenceDataGenerator(tf.keras.utils.Sequence):

    def __init__(self, X_data, y_data, batch_size, seq_length, img_dims, n_classes=2, shuffle=True):
        self.X_data = X_data         # Full array of frames
        self.y_data = y_data         # Full array of labels
        self.batch_size = batch_size
        self.seq_length = seq_length
        self.img_height, self.img_width, self.channels = img_dims
        self.n_classes = n_classes
        self.shuffle = shuffle

        self.total_samples = len(self.X_data) - self.seq_length + 1
        self.indices = np.arange(self.total_samples)

        self.on_epoch_end()

    def __len__(self):
        # Number of batches per epoch
        return int(np.floor(self.total_samples / self.batch_size))

    def __getitem__(self, index):
        # 'index' is the batch number (e.g., 0, 1, 2, ...)
        batch_start_indices = self.indices[index * self.batch_size : (index + 1) * self.batch_size]
        X_batch, y_batch = self.__data_generation(batch_start_indices)
        return X_batch, y_batch

    def on_epoch_end(self):
        # Shuffle indices at the end of each epoch
        if self.shuffle:
            np.random.shuffle(self.indices)

    def __data_generation(self, batch_start_indices):
        # Create empty arrays to hold one batch of sequence data
        X_batch = np.empty((self.batch_size, self.seq_length, self.img_height, self.img_width, self.channels), dtype=np.float32)
        y_batch = np.empty((self.batch_size), dtype=int)

        for i, start_index in enumerate(batch_start_indices):
            end_index = start_index + self.seq_length
            X_batch[i] = self.X_data[start_index : end_index]
            y_batch[i] = self.y_data[end_index - 1]

        return X_batch, y_batch

# Model Definition
def build_cnn_lstm_model(seq_length, img_height, img_width, channels):
    """Builds the CNN-LSTM architecture."""

    cnn_base = Sequential([
        Input(shape=(img_height, img_width, channels), name="cnn_input"),
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5)
    ], name="cnn_base")

    sequence_input = Input(shape=(seq_length, img_height, img_width, channels), name="video_input")
    encoded_frames = TimeDistributed(cnn_base)(sequence_input)
    lstm_out = LSTM(64)(encoded_frames)

    output = Dense(1, activation='sigmoid', name="fatigue_output")(lstm_out)

    # 'output' is now a KerasTensor, not a Layer object
    cnn_lstm_model = Model(inputs=sequence_input, outputs=output)

    cnn_lstm_model.compile(optimizer='adam',
                           loss='binary_crossentropy',
                           metrics=['accuracy'])

    return cnn_lstm_model

# Main Execution

X_full, y_full = load_frames_from_folders(
    DATA_DIR, CLASSES, IMG_HEIGHT, IMG_WIDTH, MAX_FRAMES_PER_CLASS
)

if X_full.shape[0] == 0:
    print("Error: No data was loaded.")
else:
    print(f"Data loaded successfully. Shapes: X={X_full.shape}, y={y_full.shape}")

    # 5.2. Split frame arrays
    X_train_frames, X_test_frames, y_train_frames, y_test_frames = train_test_split(
        X_full, y_full, test_size=0.2, random_state=42, stratify=y_full
    )

    print("Deleting full data array to free RAM...")
    del X_full
    del y_full

    # Instantiate the Data Generators
    img_dims = (IMG_HEIGHT, IMG_WIDTH, CHANNELS)

    print("Creating data generators...")
    train_generator = SequenceDataGenerator(
        X_train_frames, y_train_frames, BATCH_SIZE, SEQ_LENGTH, img_dims, n_classes=2, shuffle=True
    )

    validation_generator = SequenceDataGenerator(
        X_test_frames, y_test_frames, BATCH_SIZE, SEQ_LENGTH, img_dims, n_classes=2, shuffle=False
    )

    print(f"Total training sequences: {train_generator.total_samples}")
    print(f"Total validation sequences: {validation_generator.total_samples}")

    # Build the model
    model = build_cnn_lstm_model(SEQ_LENGTH, IMG_HEIGHT, IMG_WIDTH, CHANNELS)
    model.summary()

    # Define Callbacks
    checkpoint_path = "best_fatigue_model.h5"
    model_checkpoint = ModelCheckpoint(
        filepath=checkpoint_path,
        save_best_only=True,
        monitor='val_accuracy',
        mode='max',
        verbose=1
    )

    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=5, # Using the shorter patience
        restore_best_weights=True,
        verbose=1
    )

    # Train the model using the generators
    print("--- Starting Model Training ---")

    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=validation_generator,
        callbacks=[model_checkpoint, early_stopping]
    )

    print("--- Training Finished ---")

    # Save the final model
    model.save("final_fatigue_cnn_lstm_model.h5")
    print(f"Model saved to 'final_fatigue_cnn_lstm_model.h5'")
    print(f"Best model (based on val_accuracy) saved to '{checkpoint_path}'")
```

##Bits implementation

```python
import tensorflow as tf
import os
from google.colab import files

def process_and_download_model(model, model_name):
    print(f"Processing model: {model_name}...")

    if model is None:
        print(f"Error: The model variable '{model_name}' is empty. Make sure you have run the training first.")
        return

    try:
        # BIT APPLICATION (QUANTIZATION)
        print("   Starting quantization (8-bit reduction)...")
        converter = tf.lite.TFLiteConverter.from_keras_model(model)

        converter.optimizations = [tf.lite.Optimize.DEFAULT]

        converter.target_spec.supported_ops = [
            tf.lite.OpsSet.TFLITE_BUILTINS,
            tf.lite.OpsSet.SELECT_TF_OPS
        ]

        # Convert
        tflite_quant_model = converter.convert()

        # Save file
        filename = f"{model_name}_bits.tflite"
        with open(filename, 'wb') as f:
            f.write(tflite_quant_model)

        size_mb = os.path.getsize(filename) / (1024 * 1024)
        print(f"   Success! Quantized model generated: {filename} ({size_mb:.2f} MB)")

        print("   Downloading file...")
        files.download(filename)

    except Exception as e:
        print(f"   Technical Error: {str(e)}")
try:
    process_and_download_model(model_mobilenet, "MobileNetV2_Fatigue")
except NameError:
    print("Variable 'model_mobilenet' not found. Make sure you have run the MobileNet training cell.")
try:
  process_and_download_model(cnn_lstm_model, "CNNLSTM_Fatiga")
except NameError:
  pass
```
